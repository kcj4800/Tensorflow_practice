```python
import pandas as pd
import numpy as np

raw = pd.read_table('drive/MyDrive/Colab Notebooks/shopping.txt', names = ['rating', 'review'])
raw2 = pd.read_table('shopping.txt', names = ['rating', 'review'])

# np.where( a > 10, x, y) : a가 10보다 크면 x, 그렇지 않으면 y
raw['label'] = np.where(raw['rating'] > 3, 1, 0 )
raw2['label'] = np.where(raw['rating'] > 4, 1, 0)

print(raw)
print(raw2)
```

```python
# 데이터 이쁘게 하는 작업 - 특수문자 및 필요없는 문자 제거

raw['review'] = raw['review'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣0-9 ]', '')
raw2['review'] = raw2['review'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣0-9a-zA-Z ]', '')

# raw와 raw2모두 공백인 열은 없으므로, 패스
# print(raw.isnull().sum())
# print(raw2.isnull().sum())

# 중복 제거 drop_duplicates : https://wikidocs.net/154060
# subset : 중복값을 검사할 열 입니다. 기본적으로 모든 열을 검사합니다.
# keep : {first / last} 중복제거를할때 남길 행입니다. first면 첫값을 남기고 last면 마지막 값을 남깁니다.
# inplace : 원본을 변경할지의 여부입니다.
# ignore_index : 원래 index를 무시할지 여부입니다. True일 경우 0,1,2, ... , n으로 부여됩니다.

a = len(raw)
b = raw['review'][13]
print(b)
raw.drop_duplicates( subset = ['review'], inplace = True) # 
print(a, len(raw), a == len(raw)) # a == len(raw) : True 이므로, 내용이 중복되는 review는 없었나보다.

# 반복되는 단어 중복 제거( 예 : ㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋㅋ )
# lambda 매개변수 : 표현식
import itertools
raw['review'] = raw['review'].map(lambda x: ''.join(ch for ch, _ in itertools.groupby(x)))
print(raw['review'][13], '\n', b, '\n', len(b), ' => ', len(raw['review'][13]), sep='') # 54 -> 52
print(len(raw))

# Bag of words : 유니크_단어
unique_words = raw['review'].tolist()
print(unique_words[0:100])
unique_words = ''.join(unique_words)
print(unique_words[0:100])
unique_words = list(set(unique_words))
print(unique_words[0:100])
unique__words = sorted(unique_words)
unique_words.sort()
print(unique__words[0:100])
print(unique_words[0:100])
```

```python
# tensorflow.keras.preprocessing.text - Tokenizer
from tensorflow.keras.preprocessing.text import Tokenizer

# char_level = True : 글자단위를 정수로 바꾼다. char_level = False : 단어단위를 정수로 바꾼다.
# 1000 : 앞에서부터 1000개의 인덱스만 tokenizing한다. 최다 빈출 단어들로만 tokenizing
tokenizer = Tokenizer(1000, char_level = True, oov_token = '<OOV지롱>') # Out Of Vocabullary
str_list = raw['review'].tolist()
tokenizer.fit_on_texts(str_list)
# tokenizer.word_index : text_to_num 딕셔너리를 자동으로 생성해줌.
print(tokenizer.word_index)
tknizer = Tokenizer(1000, char_level = True, oov_token = '<OOV>')
tknizer.fit_on_texts(unique_words)
print(tokenizer.word_index)
print(tknizer.word_index)
# 자주쓰이는 단어 순으로 인덱싱 되므로, 1회 이하 출현단어는 index에서 삭제하는 방법을 쓰기도 한다.
# train / val 구분없이 tokenizer에 다 집어 넣었으나, train만 따로 tokenizing 할수있다.

print(str_list[0:20])

# tokenizing 된 tokenizer 변수를 이용하여 str_list를 texts_to_seqeunces( '배공빠르고 굿' => [40, 323, 87, 86, 5, 2, 344] ) 하여 train_seq에 저장해준다.
train_seq = tokenizer.texts_to_sequences(str_list) 
train_seq_unique = tknizer.texts_to_sequences(unique_words)

print(train_seq[0:20])
print(unique_words[0:20])
print(train_seq_unique[0:20])
# unique_words를 tokenizing한 tknizer의 경우 중복이 제거되고, 순서가 .sort() 되어 있어 빈도수를 이용해 index에서 삭제하는 방법을 쓰기 어렵다.

Y = raw['label'].tolist()
print(Y[0:10])
```
